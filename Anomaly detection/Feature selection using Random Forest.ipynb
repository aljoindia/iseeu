```python
# creating random forest model
rf_model = RandomForestClassifier(n_estimators=50)
```

```python
# converting all non benign labels to MALICIOUS
labels = ["BENIGN" if label == "BENIGN" else "MALICIOUS" for label in pre_pca_ids_dataset[' Label']]
```

```python
# fitting random forest model
rf_model.fit(pre_pca_ids_dataset.drop([' Timestamp', ' Label'], axis=1), labels)
```

```python
# pre_pca_ids_dataset.columns
feature_importance = pd.DataFrame(np.cumsum(rf_model.feature_importances_), columns=['importance'])
feature_importance['features'] =  rf_model.feature_names_in_
```
